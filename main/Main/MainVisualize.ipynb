## Setup

```python
#setup the visualizer. Bellow you can add new stats with add_stats 

#load in the visualizer
visualizer = vz.MNISTStatsVisualizer()
#Load in the dataset
visualizer.load_mnist_test_data("mnist_default")
```

### When adding data use stat type 1(has accuracy data) except for autoencoders type 2

```python
#load data to visualizer

# Phase1:
visualizer.add_stats("SimpleShapeNeuralNetwork", hf.load_JSON_stats("SimpleShapeNeuralNetwork_090220252.json"), stats_type=1)
visualizer.add_stats("BasicNeuralNetwork", hf.load_stats("NeuralNetwork002.pth"), stats_type=1)
visualizer.add_stats("CONVNet", hf.load_stats("CONVNet001.pth"), stats_type=1)

# Phase2
visualizer.add_stats("SimpleAutoencoder", hf.load_stats("SimpleAutoencoder001.pth"), stats_type=2)
visualizer.add_stats("ClassifierOfEncodedData", hf.load_stats("ClassifierOfEncodedData001.pth"), stats_type=1)

#Phase3
visualizer.add_stats("OldTransformer", hf.load_stats("VisualTransformerSevenPatchSize001.pth"), stats_type=1)
visualizer.add_stats("VisualTransformer", hf.load_stats("VisualTransformerTest001.pth"), stats_type=1)
visualizer.add_stats("VisualTransformer Test", hf.load_stats("VisualTransformer001.pth"), stats_type=1)
visualizer.add_stats("AltVisualTransformer GELU", hf.load_stats("AltVisualTransformerGELU001.pth"), stats_type=1)
visualizer.add_stats("AltVisualTransformer NoPatch Adam", hf.load_stats("AltVisualTransformerGELUNoPathchAdam001.pth"), stats_type=1)
visualizer.add_stats("AltVisualTransformer NoPatch", hf.load_stats("VisualTransformer001.pth"), stats_type=1)
visualizer.add_stats("ShapeTransformer small", hf.load_stats("ShapeTransformerSmallSDG001.pth"), stats_type=1)
visualizer.add_stats("ShapeTransformer big", hf.load_stats("ShapeTransformerBigSDG001.pth"), stats_type=1)



phase_one_list = ["SimpleShapeNeuralNetwork", "BasicNeuralNetwork", "CONVNet"]
phase_two_list = [ "SimpleAutoencoder", "ClassifierOfEncodedData"]
phase_three_list = [ "OldTransformer", "VisualTransformer", "VisualTransformer Test","AltVisualTransformer GELU", \
                    "AltVisualTransformer NoPatch Adam", "AltVisualTransformer NoPatch", "ShapeTransformer small", "ShapeTransformer big"]
```

## visualizations

```python
# mistake matries

viz_list = phase_one_list + [phase_two_list[1]]

for model in viz_list:
    visualizer.plot_mistake_matrix(model)

```

```python
# Histograms



visualizer.plot_double_histogram(phase_one_list) # histogram becomes to busy with three models!!


visualizer.plot_double_histogram([ "SimpleShapeNeuralNetwork" ,"ClassifierOfEncodedData"])
visualizer.plot_double_histogram([ "BasicNeuralNetwork", "CONVNet"])
```

```python
#visualize phase one and two
visualizer.plot_accuracy(phase_one_list + [phase_two_list[1]])

#visualize phase three in two parts
visualizer.plot_accuracy( [phase_three_list[i] for i in [0,1,2,3]] )
visualizer.plot_accuracy( [phase_three_list[i] for i in [4,5,6,7]]   )

#Shape based compred with the best 
visualizer.plot_accuracy( [ phase_one_list[0] ,   phase_three_list[2] , phase_three_list[7] ]   )

#Best of all phases

visualizer.plot_accuracy( [ phase_one_list[1] , phase_two_list[1]  , phase_three_list[2] , phase_three_list[7] ]   )


```

```python
#visualize phase one and two
visualizer.plot_loss(phase_one_list + phase_two_list)

#visualize phase three in two parts
visualizer.plot_loss( [phase_three_list[i] for i in [0,1,2,3]] )
visualizer.plot_loss( [phase_three_list[i] for i in [4,5,6,7]]   )

#Shape based compred with the best 
visualizer.plot_loss( [ phase_one_list[0] ,   phase_three_list[2] , phase_three_list[7] ]   )

#Best of all phases

visualizer.plot_loss( [ phase_one_list[1] , phase_two_list[1]  , phase_three_list[2] , phase_three_list[7] ]   )

```
